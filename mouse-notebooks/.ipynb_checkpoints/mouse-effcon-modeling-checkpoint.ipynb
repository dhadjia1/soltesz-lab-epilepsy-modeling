```python
def acquire_info(J_filepath, parameters_filepath, norm=False):

    f = np.load(J_filepath, allow_pickle=True)
    J = f['J']
    if norm: J /= np.max(J)
    N = J.shape[0]
    f.close()

    simulation_kwargs = load_simulation_parameters(parameters_filepath)
    print(simulation_kwargs['tau'], simulation_kwargs['nstd'])
    simulation_kwargs['J'] = J
    H = simulation_kwargs['H']
    
    f = np.load(parameters_filepath)
    f.close()

    Ytruth = simulation_kwargs['Ytruth']
    currents    = solve_stochastic_euler(**simulation_kwargs)
    activations = np.tanh(currents)
    Ymodel      = np.matmul(J, activations)
    


    simulation_kwargs['H'] = np.zeros_like(simulation_kwargs['H'], dtype='float32')
    currents_naked    = solve_stochastic_euler(**simulation_kwargs)
    activations_naked = np.tanh(currents_naked)
    Ymodel_naked     = np.matmul(J, activations_naked)
    
    return {'J': J, 'H': H, 'Ytruth': Ytruth, 'Ymodel': Ymodel, 'Ymodel naked': Ymodel_naked, 'currents': currents, 'currents naked': currents_naked}


#Jfilepath           = '/mnt/f/dhh-soltesz-lab/ca1-data/ax-126-363/eff/5042020-epoch-0/training/iter-50.npz'
#parameters_filepath = '/mnt/f/dhh-soltesz-lab/ca1-data/ax-126-363/eff/5042020-epoch-0/additional-info.npz'
Jfilepath           = '/mnt/f/dhh-soltesz-lab/DG-data/control_99/eff/5042020-epoch-0/training/iter-250.npz'
parameters_filepath = '/mnt/f/dhh-soltesz-lab/DG-data/control_99/eff/5042020-epoch-0/additional-info.npz'
dg_data_control = acquire_info(Jfilepath, parameters_filepath, norm=False)

Jfilepath           = '/mnt/f/dhh-soltesz-lab/DG-data/204/eff/5042020-epoch-0/training/iter-245.npz'
parameters_filepath = '/mnt/f/dhh-soltesz-lab/DG-data/204/eff/5042020-epoch-0/additional-info.npz'
dg_data_sz = acquire_info(Jfilepath, parameters_filepath, norm=False)
```

```python
def plot_weights(J):
    j_weights    = extract_weights(J, directed=True)
    hist, edges = np.histogram(j_weights, density=True)

    plt.figure()
    plt.plot(edges[1:], hist, color='k')
    plt.yscale('log')
    plt.show()

plot_weights(dg_data_control['J'])
plot_weights(dg_data_sz['J'])
```

```python
def plot_traces(data_dict):

    Ymodel = data_dict['Ymodel']
    Ytruth = data_dict['Ytruth']
    print(Ytruth.shape)

    plt.figure()
    plt.plot(Ytruth.mean(axis=0), color='k')
    plt.plot(Ymodel.mean(axis=0), color='r')
    plt.show()

    fig, ax = plt.subplots(3,3, figsize=(12,8))
    for i in range(9):
        idx = np.random.randint(0, Ytruth.shape[0])
        x, y = int(i/3), int(i%3)
        ax[x,y].plot(Ytruth[idx,400:500], color='k')
        ax[x,y].plot(Ymodel[idx,400:500], color='r')
    plt.show()


    currents = data_dict['currents']
    H        = data_dict['H']

    fig, ax = plt.subplots(3,3, figsize=(12,8))
    for i in range(9):
        if i == 0:
            idx = 0
        else:
            idx = np.random.randint(0, Ytruth.shape[0])
        x, y = int(i/3), int(i%3)
        ax[x,y].plot(currents[idx,:250], color='r')
        ax[x,y].plot(H[idx,:250], color='k')
    plt.show()
    
    fig, ax= plt.subplots(1,2)
    ax[0].imshow(np.corrcoef(Ytruth), cmap='inferno', vmin=-1., vmax=1.)
    ax[1].imshow(np.corrcoef(Ymodel), cmap='inferno', vmin=-1., vmax=1.)
    plt.show()
    
plot_traces(dg_data_control)
plot_traces(dg_data_sz)
```

```python
from graph_utils import get_outgoing_degrees
from futils import data2percentile, get_ei


def get_top_outgoing(J, percentile):
    outgoing = get_outgoing_degrees(J, percentile)
    outgoing_percentile = data2percentile(outgoing)
    return np.where( outgoing_percentile > 0.90)[0]
def get_top_incoming(J, percentile):
    incoming = get_outgoing_degrees(J.T, percentile)
    incoming_percentile = data2percentile(incoming)
    return np.where( incoming_percentile > 0.90)[0]
    

control_outgoing_top = get_top_outgoing(dg_data_control['J'], 90)
control_incoming_top = get_top_incoming(dg_data_control['J'], 90)

sz_outgoing_top      = get_top_outgoing(dg_data_sz['J'], 90)
sz_incoming_top      = get_top_incoming(dg_data_sz['J'], 90)

control_otop_pre_ei  = np.asarray([get_ei(dg_data_control['J'][:,idx]) for idx in control_outgoing_top])
control_itop_post_ei = np.asarray([get_ei(dg_data_control['J'][idx,:]) for idx in control_incoming_top])

sz_otop_pre_ei  = np.asarray([get_ei(dg_data_sz['J'][:,idx]) for idx in sz_outgoing_top])
sz_itop_post_ei = np.asarray([get_ei(dg_data_sz['J'][idx,:]) for idx in sz_incoming_top])

control_otop_pre_exc = control_outgoing_top[np.where(control_otop_pre_ei > 1.0)[0]]
control_itop_post_exc = control_incoming_top[np.where(control_itop_post_ei > 1.0)[0]]

sz_otop_pre_exc      = sz_outgoing_top[np.where(sz_otop_pre_ei > 1.0)[0]]
sz_itop_post_exc     = sz_incoming_top[np.where(sz_itop_post_ei > 1.0)[0]]

print(control_otop_pre_ei.min(), sz_otop_pre_ei.min())

plt.figure()
plt.hist([control_otop_pre_ei, sz_otop_pre_ei], color=['k', 'r'], density=True)
plt.show()

plt.figure()
plt.hist([control_itop_post_ei, sz_itop_post_ei], color=['k', 'r'], density=True)
plt.show()
```

```python
def run_hoa(J, idxsA, idxsB, threshold, motif, state, shuffle=False, pos=True):
    from graph_utils import generate_snap_unweighted_graph, run_higher_order_analysis
    
    idxsA, idxsB = list(idxsA), list(idxsB)
    idxs = idxsA + idxsB
    cwd = os.getcwd()
    if shuffle:
        J = permute_weights(J, seed=shuffle)
    graph_output_filepath = cwd+'/%s.txt' % state
    Jth = jthreshold(J,threshold,above=True, pos=pos,binarized=True)
    Jth_reduced = Jth[idxs,:][:,idxs]

#     Jth_reduced[:len(idxsA)][:len(idxsA)] = 0
#     Jth_reduced[len(idxsA):][len(idxsA):] = 0
#     Jth_reduced[:len(idxsA)][len(idxsA):] = 0
    
    generate_snap_unweighted_graph(Jth_reduced.T,directed=True,save=graph_output_filepath)
    hoa_output_filepath  = cwd + '/hoa-%s.txt' % state    
    run_higher_order_analysis(graph_output_filepath, hoa_output_filepath, cwd, motif)
  
#sender_thalamus sender_anti
# m5 and bifan dg
motif = 'm5'
pos   = True

print(len(control_otop_pre_exc)+ len(control_itop_post_exc))
print(len(sz_otop_pre_exc)+ len(sz_itop_post_exc))

run_hoa(dg_data_control['J'], control_otop_pre_exc, control_itop_post_exc, 90, motif, 'baseline', shuffle=0, pos=pos)
run_hoa(dg_data_sz['J'], sz_otop_pre_exc, sz_itop_post_exc, 90, motif, 'presz', shuffle=0, pos=pos)
```

```python
def run_hoa2(J, threshold, motif, state, pos=True):
    from graph_utils import generate_snap_unweighted_graph, run_higher_order_analysis
    cwd = os.getcwd()
    graph_output_filepath = cwd+'/%s.txt' % state
    Jth = jthreshold(J,threshold,above=True, pos=pos,binarized=True)
    
    generate_snap_unweighted_graph(Jth.T,directed=True,save=graph_output_filepath)
    hoa_output_filepath  = cwd + '/hoa-%s.txt' % state    
    run_higher_order_analysis(graph_output_filepath, hoa_output_filepath, cwd, motif)

# ca1 m1 is pretty big deal
motif='bifan'   
pos=True
print(len(dg_data_control['J']), len(dg_data_sz['J']))

run_hoa2(dg_data_control['J'], 90., motif, 'baseline', pos=pos)
run_hoa2(dg_data_sz['J'], 90., motif, 'presz', pos=pos)
```
